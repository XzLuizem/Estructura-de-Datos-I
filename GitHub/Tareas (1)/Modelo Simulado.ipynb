# Ejemplo 1

```python
import random

def run_simulation(num_steps: int):
    """
    Ejecuta una simulación simple.

    Args:
        num_steps: El número de pasos a simular.

    Returns:
        El estado final de la simulación.
    """
    # Estado inicial
    state = 0

    # Bucle de simulación
    for step in range(num_steps):
        # Actualización del estado (ejemplo simple: sumar un número aleatorio)
        change = random.randint(-1, 1)
        state += change

        # Opcional: Imprimir el estado en cada paso
        # print(f"Paso {step}: Estado = {state}")

    return state

# Parámetros de la simulación
NUMBER_OF_STEPS = 10

# Ejecutar la simulación
final_state = run_simulation(NUMBER_OF_STEPS)

# Imprimir el resultado final
print(f"Estado final de la simulación: {final_state}")
```

# Ejemplo 2

```python
import random

def simple_random_walk(steps: int, start_position: int = 0):
    """
    Simula un paseo aleatorio unidimensional.

    Args:
        steps: El número de pasos en el paseo.
        start_position: La posición inicial.

    Returns:
        La posición final después de los pasos.
    """
    position = start_position
    for _ in range(steps):
        move = random.choice([-1, 1])  # Moverse a la izquierda (-1) o derecha (1)
        position += move
    return position

# Parámetros para el nuevo ejemplo
NUMBER_OF_STEPS_2 = 20
STARTING_POINT = 5

# Ejecutar la segunda simulación
final_position = simple_random_walk(NUMBER_OF_STEPS_2, STARTING_POINT)

# Imprimir el resultado
print(f"Posición final del paseo aleatorio: {final_position}")
```
