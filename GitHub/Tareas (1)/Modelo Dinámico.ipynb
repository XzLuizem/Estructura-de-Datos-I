# Ejemplo 1

```python
# Lista donde se pueden agregar o eliminar elementos.

mi_lista = [] # Inicializa una lista vacía

# Agregar elementos a la lista
mi_lista.append(5)
mi_lista.append(15)
mi_lista.append(25)

print("Lista después de agregar elementos:")
print(mi_lista)

# Acceder a elementos por índice
print("\nPrimer elemento:", mi_lista[0])
print("Segundo elemento:", mi_lista[1])
print("Tercer elemento:", mi_lista[2])

# Eliminar un elemento
mi_lista.remove(15)
print("\nLista después de eliminar el 15:")
print(mi_lista)
```

# Ejemplo 2

```python
class Nodo:
    """Representa un nodo individual en la lista enlazada."""

    def __init__(self, dato):
        """Inicializa un nuevo nodo con el dato proporcionado."""
        self.dato = dato
        self.siguiente = None


class ListaEnlazada:
    """Representa una lista enlazada simple."""

    def __init__(self):
        """Inicializa una lista enlazada vacía."""
        self.cabeza = None

    def agregar_al_final(self, dato):
        """Agrega un nuevo nodo al final de la lista."""
        nuevo_nodo = Nodo(dato)
        if self.cabeza is None:
            self.cabeza = nuevo_nodo
            return
        ultimo_nodo = self.cabeza
        while ultimo_nodo.siguiente:
            ultimo_nodo = ultimo_nodo.siguiente
        ultimo_nodo.siguiente = nuevo_nodo

    def imprimir_lista(self):
        """Imprime los elementos de la lista."""
        nodo_actual = self.cabeza
        while nodo_actual:
            print(nodo_actual.dato, end=" -> " if nodo_actual.siguiente else "")
            nodo_actual = nodo_actual.siguiente
        print(" End")


# Ejemplo de uso:
if __name__ == "__main__":
    mi_lista = ListaEnlazada()
    mi_lista.agregar_al_final(10)
    mi_lista.agregar_al_final(20)
    mi_lista.agregar_al_final(30)

    print("Lista enlazada:")
    mi_lista.imprimir_lista()
```
