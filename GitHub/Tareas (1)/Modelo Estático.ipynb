# Ejemplo 1

```python
class Config:
    """
    Clase que representa un modelo estático con configuración.
    """

    # Variables estáticas (constantes o configuraciones)
    API_KEY = "qwerty"
    DATABASE_URL = "ArmandoTorrez.Rojas.com"

    @staticmethod
    def get_api_key():
        """
        Método estático para obtener la clave de API.
        """
        return Config.API_KEY

    @staticmethod
    def get_database_url():
        """
        Método estático para obtener la URL de la base de datos.
        """
        return Config.DATABASE_URL


# Ejemplo de uso del modelo estático
if __name__ == "__main__":
    print(f"Clave de API: {Config.get_api_key()}")
    print(f"URL de la base de datos: {Config.get_database_url()}")
```

```python
class Constants:
    """
    Clase estática para agrupar constantes relacionadas.
    """
    MAX_USERS = 80
    DEFAULT_TIMEOUT = 30  # Segundos
    ADMIN_EMAIL = "luism.jim7@gmail.com"
    VERSION = "1.35"

# Ejemplo de uso de las constantes
if __name__ == "__main__":
    print(f"Máximo número de usuarios: {Constants.MAX_USERS}")
    print(f"Tiempo de espera por defecto: {Constants.DEFAULT_TIMEOUT} segundos")
    print(f"Correo electrónico del administrador: {Constants.ADMIN_EMAIL}")
    print(f"Versión de la aplicación: {Constants.VERSION}")
```
