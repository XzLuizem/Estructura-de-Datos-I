# Ejemplo 1

```python
class PersistentCounter:
  """A simple counter with persistent state."""

  def __init__(self, initial_value=0):
    """Initializes the counter."""
    self._value = initial_value

  def increment(self):
    """Increments the counter by 1."""
    self._value += 1

  def get_value(self):
    """Returns the current value of the counter."""
    return self._value

# Example usage:
counter = PersistentCounter(5)
counter.increment()
print(f"Current value: {counter.get_value()}")

counter.increment()
print(f"Current value: {counter.get_value()}")
```

# Ejemplo 2

```python
import random

class PersistentRandomWalk:
    """A simple random walk with persistent state."""

    def __init__(self, initial_position=0):
        """Initializes the random walk."""
        self._position = initial_position

    def step(self):
        """Takes a random step (+1 or -1)."""
        direction = random.choice([-1, 1])
        self._position += direction

    def get_position(self):
        """Returns the current position of the random walk."""
        return self._position

# Example usage:
walk = PersistentRandomWalk()
for _ in range(10):
    walk.step()

print(f"Posición final del paseo aleatorio: {walk.get_position()}")
```
